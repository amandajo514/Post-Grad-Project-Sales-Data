#### Part IV: Modeling

```python
# Map 'Gender' column using lambda function
df['Gender'] = df['Gender'].map(lambda x: 1 if x == 'Male' else 2 if x == 'Female' else x)
```

```python
# Check unique values in 'Gender' column after mapping
unique_genders_after_mapping = df['Gender'].unique()
print("\nUnique Genders (After Mapping):\n", unique_genders_after_mapping)
```

```python
# Drop unnecessary columns for modeling
columns_to_drop = ['Date', 'Customer ID', 'Age Group']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
# Check the modified dataframe
print("\nModified DataFrame:\n", df.head())
```

```python
# Baseline
gender_proportion = df['Gender'].value_counts(normalize=True)
print("\nBaseline Gender Proportion:\n", gender_proportion)
```

##### ** The baseline is 51%

```python
# Create X & y
X = df.drop(columns='Gender')
y = df['Gender']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Drop the 'Total Amount Group' column if it still exists
X_train = X_train.drop(columns='Total Amount Group', errors='ignore')
X_test = X_test.drop(columns='Total Amount Group', errors='ignore')
```

```python

# StandardScaler
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

```python
# Istantiate
```

```python
knn = KNeighborsClassifier()
```

```python
# Fit model
```

```python
knn.fit(X_train_sc,y_train)
```

```python
#Accuracy Score
```

```python
knn.score(X_train_sc, y_train)
```

```python
knn.score(X_test_sc,y_test)
```

```python
# Confusion Matrix
```

```python
ConfusionMatrixDisplay.from_estimator(knn, X_test_sc, y_test, cmap = 'Blues', display_labels = ["False", "True"]);
```

```python
Gender = []
```

```python
for k in range (3,32,2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_sc,y_train)
    acc_train=knn.score(X_train_sc, y_train)
    acc_test=knn.score(X_test_sc,y_test)
    Gender.append({'k':k,'Training Accuracy':acc_train,'Test Accuracy':acc_test})

df_Gender = pd.DataFrame(Gender)
df_Gender
```

```python
sns.lineplot(data = df_Gender, x = 'k', y = 'Training Accuracy', c='pink', label='Train')
sns.lineplot(data = df_Gender, x = 'k', y = 'Test Accuracy', c='blue', label='Test')
plt.xticks(range(3, 32, 2));
```
